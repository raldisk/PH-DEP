# 01 — Fixture EDA

Explore the pre-recorded DataCamp fixture datasets before running the live pipeline.

```python
import json
from pathlib import Path
import pandas as pd
from collections import Counter

FIXTURES = Path("fixtures")
```

## WW + US Trends — intersection analysis

```python
ww = json.load(open(FIXTURES / "WWTrends.json"))
us = json.load(open(FIXTURES / "USTrends.json"))

ww_names = {t["name"] for t in ww[0]["trends"]}
us_names = {t["name"] for t in us[0]["trends"]}
common   = ww_names & us_names

print(f"WW trends:  {ww_names}")
print(f"US trends:  {us_names}")
print(f"Common: {common}")
```

## Tweet-level EDA

```python
tweets = json.load(open(FIXTURES / "WeLoveTheEarth.json"))
df = pd.DataFrame([{"id": t["id"], "text": t["text"], "lang": t["lang"]} for t in tweets])

print(f"Total tweets: {len(df)}")
print(f"Language breakdown:\n{df['lang'].value_counts()}")
print(f"\nSample texts:")
for _, row in df.head(5).iterrows():
    print(f"  [{row['lang']}] {row['text'][:80]}...")
```

## Hashtag frequency

```python
import re
hashtags = []
for t in tweets:
    hashtags.extend(re.findall(r"#(\w+)", t["text"], re.UNICODE))

print("Top hashtags:")
for tag, count in Counter(hashtags).most_common(10):
    print(f"  #{tag}: {count}")
```
