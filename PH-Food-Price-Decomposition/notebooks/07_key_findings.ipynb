## Finding 1 — Rice structural doubling

National retail price of well-milled rice was approximately **₱28/kg in 2010**. By February 2026 it reached **₱53.54/kg** — a 91% increase over 15 years. The Rice Tariffication Law (RA 11203, March 2019) temporarily reversed the upward trend as imports surged from Vietnam, Thailand, and Myanmar, but supply-demand pressures resumed from 2022 onwards.

*Source: PSA Price Situationer, 2000–2026*

---

## Finding 2 — The Onion Crisis (January 2023)

The STL residual component shows the January 2023 onion price spike was **not seasonal** — it was a pure supply disruption. Onions hit ₱700/kg nationally, briefly exceeding chicken prices. The shock was amplified by concurrent smuggling enforcement actions reducing legitimate import volumes against a backdrop of poor domestic harvests.

*Source: PSA Price Situationer; STL residual z-score > 4.5σ*

---

## Finding 3 — Diesel is the food price transmission mechanism

Cross-correlation analysis shows diesel price changes appear in tomato and vegetable prices at a **6–8 week lag**. A ₱5/litre diesel increase translates to an average ₱3–5/kg increase in vegetables within 2 months. The Philippines' dependence on road and marine logistics for food distribution makes diesel the dominant cost pass-through channel.

*Source: DOE Oil Monitor + PSA Price Situationer cross-correlation*

---

## Finding 4 — August–September CPI spike is predominantly seasonal

STL seasonal components show that ~70% of the annual August–September food CPI spike is explained by the seasonal component — not by structural trends or shocks. Driven by rice harvest timing (Q4 surplus hasn't arrived yet) and back-to-school demand for packed lunches and canteen supplies.

*Source: PSA OpenSTAT CPI monthly series*

---

## Finding 5 — Vegetable seasonality is 3× stronger than rice seasonality

The seasonal amplitude (peak-to-trough ratio) for onions and tomatoes is 2.8–3.2× that of rice. Rice seasonality has weakened substantially since 2019 as RTL-driven imports smooth domestic harvest cycles. Vegetables remain highly seasonal because domestic production dominates and cold-chain infrastructure is limited.

*Source: STL seasonal component analysis*

---

## Finding 6 — Ukraine war was the largest correlated multi-commodity shock

The February 2022 Russia invasion of Ukraine produced simultaneous z-score spikes in cooking oil (sunflower) and wheat that were larger than any single-commodity shock outside COVID. The Suez Canal blockage (March 2021) was a precursor — cooking oil residuals spiked 14 months before the Ukraine war compounded the disruption.

*Source: STL residual analysis + World Bank Pink Sheet global prices*

---

## Finding 7 — COVID was a distribution shock, not a supply shock

The STL trend components for rice and meat show **no structural break** in 2020 — prices spiked due to panic buying and logistics disruption, then normalized within 4 months. The 2023 onion crisis and 2022 cooking oil shock, by contrast, show genuine structural residuals persisting for 6–12 months.

*Source: STL trend + residual decomposition 2020–2023*

---

## Finding 8 — February 2026 headline inflation 2.4% — food remains the driver

PSA official report (February 2026): headline CPI at 2.4%. Food and non-alcoholic beverages remain the primary driver. Within food, rice and vegetables account for the largest share of the headline contribution, consistent with the structural price trends identified in this analysis.

*Source: PSA official February 2026 inflation report*
