## Key Findings Summary

### Finding 1 — 2023 is a Gini milestone
The national Gini coefficient dropped to **0.387** in 2023 — the first time in recorded history it fell below 0.40, the World Bank threshold for high-inequality countries. This is a structural shift, not a cyclical improvement.

### Finding 2 — BARMM's unprecedented poverty reduction
BARMM's poverty incidence fell from **61.2% (2021) to 37.2% (2023)** — a 24-point drop, the largest regional improvement nationally in any 2-year period on record.

### Finding 3 — COVID reversed the usual Metro Manila resilience narrative
During the pandemic, NCR poverty *increased* while BARMM continued to improve. OFW remittances acted as a macro shock absorber in poorer regions while service-economy workers in NCR bore the brunt.

### Finding 4 — Majority of LGUs now below 20% poverty
**54.7% of the 1,611 LGUs** covered by the 2023 SAE have poverty incidence at or below 20% — a majority for the first time.

### Finding 5 — Farmers and fisherfolk remain structurally poor
Despite national improvements, occupational poverty remains highest among farmers (~30%) and fisherfolk (~30%). Geography of poverty is shifting but occupational poverty is not.

### Finding 6 — NCR-BARMM income gap is a 3.6x multiplier
Mean family income in NCR (₱580K/year) is **3.6 times** the BARMM mean (₱160K/year). The gap is narrowing but remains the largest regional income disparity in the country.

### Finding 7 — Food expenditure share rises as income falls
Bottom quintile households spend **~57%** of total expenditure on food vs ~38% for the top quintile — the classic Engel's Law pattern, confirmed by FIES 2023.

### Finding 8 — GRDP growth and poverty reduction are weakly correlated
Pearson r between regional GRDP growth and poverty reduction is weak and not statistically significant at p<0.05 — indicating that aggregate growth alone does not explain poverty reduction. Targeted transfers and employment programs matter more at the regional level.
